## support vector machines

Support vector machines or SVM as its called is a supervised learning model. Its one of the most preffered algohoritms in machine learning due to its accuracy compered to computational power needed. SVM is mostly used for classification, but it can also be used in regression. SVM is built around the idea of hyperplanes, so what are these hyperplanes? 



If we take a look at a $p$-dimensional space, hyperplane is then a affine subspace of  $p-1$. This means if we take a look at two and three dimensions a hyperplane would first be a straight line and then a simple plane. Mathmatecliy we can describe a hyperplane in two dimensions as 

$$
b+w_1x_1+w_2x_2=0,
$$

where $b$ is the intercept and $w_1$ and $w_2$ define the elements of a vector orthogonal to the line 
$b+w_1x_1+w_2x_2=0$. 
In two dimensions we define the vectors $\boldsymbol{x} =[x1,x2]$ and $\boldsymbol{w}=[w1,w2]$. 
We can then rewrite the above equation as

$$
\boldsymbol{x}^T\boldsymbol{w}+b=0.
$$
